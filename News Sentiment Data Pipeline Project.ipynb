```python
def fetch_news(keyword):
    page_size = 10
    api_key= "e0a6dff986854240b7cec38e76817d60"
    url = f"https://newsapi.org/v2/everything?q={keyword}&language=en&pageSize={page_size}&apiKey={api_key}"
    response =requests.get(url)
    data= response.json()
    return data["articles"]

articles = fetch_news("stocks") 
#stock market

#s= fetch_news("Tesla")
#print(s)

def get_tags(text):
    tags = []
    key={
        "earnings": ["earnings", "quarterly results", "revenue"],
        "ipo": ["IPO", "initial public offering"],
        "merger": ["merger", "acquisition", "buyout"],
        "layoffs": ["layoff", "job cut", "downsizing"]
    }
    for tag,words in key.items():
        if any (word in text for word in words):
            tags.append(tag)
    return tags

            
def analyze_sentiment(title, description):
    """
    Combine title and description and use TextBlob to determine sentiment.
    Returns: sentiment label (Positive/Negative/Neutral), polarity score.
    """
    full_text = (title or "") + ". " + (description or "")
    blob = TextBlob(full_text)
    score = blob.sentiment.polarity
    
    if score > 0.1:
        sentiment = "Positive"
    elif score < -0.1:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    
    return sentiment, score, full_text


def fetch_news_with_sentiment(articles):
    result = []
    for article in articles:
        title = article.get("title", "")
        description = article.get("description", "")
        publishedAt = article.get("publishedAt", "")
        url = article.get("url", "")
        #source_name = article['source']['name']
        source_name = article.get('source', {}).get('name', '')

        sentiment, score, full_text = analyze_sentiment(title, description)
        
        
        #tags = get_tags(title + " " + description)
        
        tags = get_tags((title or "") + " " + (description or ""))


        result.append({
            "title": title,
            "description": description,
            "publishedAt": publishedAt,
            "url": url,
            "sentiment": sentiment,
            "score": score,
            "source_name": source_name,
            "word_count": len(full_text.split()),
            "title_length": len(title),
            "tags":"".join(tags)
        })
    return result


final_result = fetch_news_with_sentiment(articles)
df = pd.DataFrame(final_result)

print(df.head())
print(df.info())
#df["publishedAt"] = pd.to_datetime(df["publishedAt"]).dt.strftime("%Y-%m-%d %H:%M:%S")
df["publishedAt"] = pd.to_datetime(df["publishedAt"])
print(df.info())




        
        
```
